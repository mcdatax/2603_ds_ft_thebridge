## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

````python
import math
grados = float(input("Introduce los grados para pasarlos a radianes: ")) 
radianes = math.radians(grados)
print(f'{grados} grados son {round(radianes, 5)} radianes.')

radianes = grados * math.pi / 180
print(f'{grados} grados son {round(radianes, 5)} radianes.') # Esta solución es con la fórmula del ejercicio.
````

## Ejercicio 6
Mismo programa que en 5, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
var1 = input("Introduce el primer valor a analizar: ")
var2 = input("Introduce el segundo valor a analizar: ")
var3 = input("Introduce el tercer valor a analizar: ")

todos_iguales = var1 == var2 == var3
al_menos_dos = var1 == var2 or var1 == var3 or var2 == var3

if todos_iguales:
    print(f'Los valores {var1}, {var2} y {var3} son todos iguales')
elif al_menos_dos:
    print(f'Al menos dos de los valores {var1}, {var2} y {var3} son iguales')
else:
    print(f'Los valores {var1}, {var2} y {var3} son todos diferentes')

````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10

````python
var1 = int(input("Introduce el primer valor entero: "))
var2 = int(input("Introduce el segundo valor entero: "))

suma = var1 + var2
if suma == 10:
    print(f'La suma de {var1} y {var2} es igual a 10')
elif suma > 10:
    print(f'La suma de {var1} y {var2} es mayor que 10')
else:
    print(f'La suma de {var1} y {var2} es menor que 10')

````

## Ejercicio 10
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

1. Pásalo todo a mayúsculas
2. Pásalo todo a minúsculas
3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula
4. Crea una lista dividiéndolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

````python
cadena = "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."
print(cadena.upper())
print(cadena.lower())
print(cadena.title())
lista = cadena.split()
print(lista)
cadena_reemplazada = cadena.replace(",", ";")
print(cadena_reemplazada)
print(cadena.replace("a", ""))
````
