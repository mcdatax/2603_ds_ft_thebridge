## Ejercicio 5
Realiza un conversor de grados Celsius a Fahrenheit. El conversor tiene que recibir el input del usuario y devolver la conversión.

La conversión de grados sigue la siguiente fórmula: F = (C × 9 / 5) + 32 

```python
grados = input("Introduce los grados centígrados: ")
grados_fahrenheit = (float(grados) * 9/5) + 32
print(f'{grados} grados Celsius son {grados_fahrenheit} grados Fahrenheit.')
```

```python
'''
En esta celda se pide al usuario que introduzca los grados Celsius a través de la función input(), y se guarda ese valor en la variable grados como un string. 
Luego, se convierte ese valor a un número decimal (float) para poder realizar la operación matemática de conversión a Fahrenheit, 
y finalmente se imprime el resultado utilizando una f-string para mostrar tanto los grados Celsius como su equivalente en Fahrenheit.

    Args:
        grados (str): Grados Celsius introducidos por el usuario
        grados_fahrenheit (float): Grados Fahrenheit calculados a partir de los grados Celsius

    Returns:
        float: Equivalente en grados Fahrenheit

'''
grados = input("Introduce los grados centígrados: ") # El input pregunta al usuario por los grados Celsius y lo guarda en la variable grados como un string
grados_fahrenheit = (float(grados) * 9/5) + 32 # Convertimos el string a un número decimal (float) para poder hacer la operación matemática de convertirlo a Fahrenheit
print(f'{grados} grados Celsius son {grados_fahrenheit} grados Fahrenheit.')# Imprime el resultado usando una f-string para mostrar los grados Celsius y su equivalente en Fahrenheit

```
